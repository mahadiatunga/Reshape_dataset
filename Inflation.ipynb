```python
df = pd.read_excel("C:/Users/Arshad/Documents/Salu/Data_Extract_From_World_Development_Indicators_Region.xlsx",
                   sheet_name ='Data',
                       skipfooter=4)
#skiprows=range(10)-->for skipping first 10 rows, not needed here
```

```python
df.rename (columns = {'Country Name':'Country', 'Series Name':'Series'}, inplace = True)
df.drop(['Country Code','Series Code'], axis = 1, inplace = True)
df
```

```python
df2 = pd.melt(df, id_vars=['Country','Series'],var_name='Year', value_name='value')
df2
```

```python
df2['Year']=df2['Year'].astype(str)
df2_split = df2["Year"].str.split(" ", n = 1, expand = True) 
df2['Year1']=df2_split[0]
df2['Year2']=df2_split[1]
df2.drop(columns =["Year"], inplace = True)
df2.drop(columns =["Year2"], inplace = True)
df2.rename({'Year1': 'Year'}, axis=1, inplace=True)
df2
```

```python
df3 = df2.pivot_table(index = ['Country','Year'], values = 'value',columns = 'Series', aggfunc='sum')
df3.to_csv (r'C:/Users/Arshad/Documents/Salu/df_years_series_final4.csv')
df3
```
